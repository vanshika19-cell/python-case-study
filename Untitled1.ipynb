```python
import math

# ---------- Functions for operations ----------

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b

def power(a, b):
    return a ** b

def modulus(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a % b

def factorial(n):
    if n < 0:
        return "Factorial not defined for negative numbers"
    return math.factorial(n)

def square_root(n):
    if n < 0:
        return "Square root not defined for negative numbers"
    return math.sqrt(n)


# ---------- Main menu loop ----------

while True:
    print("\n------ Advanced Calculator ------")
    print("1 Addition")
    print("2 Subtraction")
    print("3 Multiplication")
    print("4 Division")
    print("5 Power")
    print("6 Modulus")
    print("7 Factorial")
    print("8 Square Root")
    print("9 Exit")

    try:
        choice = int(input("Enter choice: "))
    except ValueError:
        print("Invalid numeric input")
        continue

    if choice == 9:
        print("Program terminated")
        break

    # operations needing two numbers
    if choice in [1, 2, 3, 4, 5, 6]:
        try:
            a = float(input("Enter first number: "))
            b = float(input("Enter second number: "))
        except ValueError:
            print("Invalid numeric input")
            continue

        if choice == 1:
            print("Result =", add(a, b))
        elif choice == 2:
            print("Result =", subtract(a, b))
        elif choice == 3:
            print("Result =", multiply(a, b))
        elif choice == 4:
            print("Result =", divide(a, b))
        elif choice == 5:
            print("Result =", power(a, b))
        elif choice == 6:
            print("Result =", modulus(a, b))

    # operations needing one number
    elif choice == 7:
        try:
            n = int(input("Enter number: "))
            print("Factorial =", factorial(n))
        except ValueError:
            print("Invalid numeric input")

    elif choice == 8:
        try:
            n = float(input("Enter number: "))
            print("Square Root =", square_root(n))
        except ValueError:
            print("Invalid numeric input")

    else:
        print("Invalid choice. Try again.")
```

```python
# Temperature Monitoring and Control System

# Function to classify temperature
def classify_temp(temp):
    if temp < 18:
        return "Cold"
    elif temp <= 28:
        return "Normal"
    else:
        return "Hot"


# Function to calculate statistics
def calculate_stats(temps):
    avg = sum(temps) / len(temps)
    highest = max(temps)
    lowest = min(temps)
    return avg, highest, lowest


# Function to display report
def display_report(temps, counts):
    avg, highest, lowest = calculate_stats(temps)

    print("\nTemperature Summary")
    print("Average Temperature =", round(avg, 2))
    print("Highest Temperature =", highest)
    print("Lowest Temperature =", lowest)

    print("\nCondition Count")
    print("Hot =", counts["Hot"])
    print("Normal =", counts["Normal"])
    print("Cold =", counts["Cold"])

    print("\nSuggested Action")

    if counts["Hot"] > 0:
        print("Fan ON")
    if counts["Cold"] > 0:
        print("Heater ON")
    if counts["Hot"] == 0 and counts["Cold"] == 0:
        print("No action required")


# Main program
def main():
    n = int(input("Enter number of readings: "))

    temps = []
    counts = {"Hot": 0, "Normal": 0, "Cold": 0}

    for i in range(n):
        temp = float(input("Enter temperature: "))
        temps.append(temp)

        category = classify_temp(temp)
        counts[category] += 1

    display_report(temps, counts)


# Run program
main()
```
